### display image in python using opencv and matplotlib

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#using opencv
img=cv2.imread('minion.jpg')
newimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
plt.imshow(newimg)
plt.show()
```
